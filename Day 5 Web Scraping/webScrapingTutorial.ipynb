```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
webpage=requests.get('https://quotes.toscrape.com/page/2/').text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
quote = soup.find_all("div",class_ ="quote")
```

```python
len(quote)
```

```python
author=[]
texts=[]
tags=[]
for i in quote:
    author.append(i.find("small").text.strip())
    texts.append(i.find("span",class_ ="text").text.strip())
    tags.append(i.find("a",class_ ="tag").text)
d = {"Author":author ,"Motivational Quote": texts, "Tag":tags}

df = pd.DataFrame(d)
```

```python
final = pd.DataFrame()
for j in range(1,11):
    url = f'https://quotes.toscrape.com/page/{j}/'
    
    webpage=requests.get(url).text

    soup=BeautifulSoup(webpage,'lxml')

    quote = soup.find_all("div",class_ ="quote")
    author=[]
    texts=[]
    tags=[]
    
    for i in quote:
        author.append(i.find("small").text.strip())
        texts.append(i.find("span",class_ ="text").text.strip())
        tag_elems = i.find_all("a", class_="tag")
        tags.append(", ".join(t.text.strip() for t in tag_elems))



        
    d = {"Author":author ,"Motivational Quote": texts, "Tag":tags}
    
    df = pd.DataFrame(d)

    final = pd.concat([final, df], ignore_index = True)

final.shape
```
