```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

# 1.ScatterPlot(numerical - numerical)

we are working/showing 3 or more parameters in the same graph thats why its multivariate analysis

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker',size='size')
```

# 2.Bar Plot(numerical - Categorical)

```python
titanic= pd.read_csv("Titanic-Dataset.csv")
```

```python
titanic.head()
```

```python
sns.barplot(x='Pclass', y='Age', data=titanic, hue="Sex",palette = "Reds")
```

# 3.BOX Plot(numerical - Categorical)

```python
sns.boxplot(x='Sex', y='Age', data=titanic, hue="Survived")
```

# 4.Dist Plot(numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived'] == 1]['Age'],hist=False)
```

# 5.HEAT MAP(Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
titanic.groupby('Pclass')['Survived'].mean()*100
```

# 6.Cluster MAP(Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```

# 7. PairPlot

```python
sns.pairplot(titanic, hue = 'Survived')
```

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue = 'species')
```

# 8. linePlot(numerical-numerical)

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
new = flights.groupby('year')['passengers'].sum().reset_index()
```

```python
sns.lineplot(data=new, x='year', y='passengers')
```

```python
flights.pivot_table(
    values='passengers',
    index='month',
    columns='year',
    observed=True
)
```

```python
sns.heatmap(flights.pivot_table(
    values='passengers',
    index='month',
    columns='year',
    observed=True
))
```

```python
sns.clustermap(flights.pivot_table(
    values='passengers',
    index='month',
    columns='year',
    observed=True
))
```
